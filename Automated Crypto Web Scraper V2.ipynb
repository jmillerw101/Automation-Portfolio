```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
import os
import time
```

```python
def automated_crypto_pull():
    url='https://coinmarketcap.com/currencies/bitcoin'
    page=requests.get(url,headers={'User-Agent':'Mozilla/5.0'})
    soup=BeautifulSoup(page.text,'html.parser')
    name=soup.select_one('h1').text
    price=soup.select_one('span[data-test="text-cdp-price-display"]').text.replace('$','').replace(',','')
    ts=datetime.now()
    df=pd.DataFrame([{'Crypto Name':name,'Price':price,'TimeStamp':ts}])
    df.to_csv(
        'Crypto_Automated_Pull.csv',
        mode='a',
        header=not os.path.exists('Crypto_Automated_Pull.csv'),
        index=False
    )
```

## Crypto Web Scraper V2 ##

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
import os
import time

def automated_crypto_pull():
    url = 'https://coinmarketcap.com/currencies/bitcoin/'

    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    page = requests.get(url, headers=headers)
    soup = BeautifulSoup(page.text, 'html.parser')

    # --- GET COIN NAME ---
    coin_name_dirty = soup.find("h1").text
    crypto_name = coin_name_dirty.replace(" price", "")

    # --- GET PRICE ---
    crypto_price = soup.find("span", class_="sc-65e7f566-0 esyGGG base-text").text
    final_price = crypto_price.replace("$", "").replace(",", "")

    date_time = datetime.now()

    row = {
        "Crypto Name": crypto_name,
        "Price": float(final_price),
        "TimeStamp": date_time
    }

    df = pd.DataFrame([row])

    file_path = r"/Users/jamillerasdf/Documents/AB Python/Crypto Web Puller/Crypto_Automated_Pull.csv"

    if os.path.exists(file_path):
        df.to_csv(file_path, mode="a", header=False, index=False)
    else:
        df.to_csv(file_path, index=False)

    print(df)

while True:
    automated_crypto_pull()
    time.sleep(10)
```
